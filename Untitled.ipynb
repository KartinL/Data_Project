```python
def float_str_to_int(str1):
    try:
        a = str(str1)
        a = a.strip( '.00' )
        a = a.strip('$')
        a = ''.join(a.split(','))
        if str1 == "$0.00":
            return 0
    #     a = a.strip(' ')
        return int(a.strip(' '))
    except Exception as e:
        print(e)
        import pdb; pdb.set_trace()
        return str1
#     return int(float(a))
```

```python
def remove(str1):
    a = str(str1)
    a = a.strip( 'NSW' )
    a = a.strip('.0')
    a = a.split('\n')
    if a == "nan":
        return a 
    return a
```
